```python
string_cols = ['Gender', 'Country','Sleep_Quality','Stress_Level', 'Health_Issues', 'Occupation']
numeric_cols = ['ID', 'Age', 'Coffee_Intake', 'Caffeine_mg','Sleep_Hours','BMI', 'Heart_Rate', 'Physical_Activity_Hours', 'Smoking', 'Alcohol_Consumption']
```

```python
def show_missing_data():
    total_rows = len(df_synthetic_coffee_health)
    print("=" * 70)
    print(f"จำนวนแถวทั้งหมดในชุดข้อมูล: {total_rows:,} แถว")
    print("=" * 70)

    print("\n--- กลุ่มตัวเลข ---")
    for col in numeric_cols:
        null_count = df_synthetic_coffee_health[col].isnull().sum()

        print(f"คอลัมน์: {col}")
        print(f"  - NULL : {null_count:,} แถว")

        print("-" * 40)

    print("\n--- กลุ่มอักษร ---")
    # ตรวจเฉพาะค่าว่างจริง (NaN / None) เท่านั้น
    missing_counts = df_synthetic_coffee_health[string_cols].isnull().sum()
    columns_with_missing_data = missing_counts[missing_counts > 0].sort_values(ascending=False)

    if not columns_with_missing_data.empty:
        print("คอลัมน์ที่มีค่าว่าง (NULL):")
        for col, count in columns_with_missing_data.items():
            print(f"  - {col}: {count:,} แถว")
    else:
        print("ไม่พบค่าว่าง (NULL) ในคอลัมน์ที่ระบุ")

    print("=" * 70)
```

# Task
Implement a machine learning pipeline using the `df` DataFrame to predict 'Is_High_Stress'. The pipeline should include preprocessing steps with `ColumnTransformer`, which applies `OneHotEncoder` to 'Sleep_Quality' and 'Health_Issues', and `StandardScaler` combined with `PolynomialFeatures` to 'Sleep_Hours', 'Caffeine_mg', and 'Heart_Rate'. After preprocessing, the pipeline should use a `LogisticRegression` model. Split the data into training and testing sets, train the pipeline, and then evaluate its performance using `classification_report` and `confusion_matrix`. Finally, extract and interpret the feature importance (coefficients) from the trained `LogisticRegression` model, considering the transformations applied by `OneHotEncoder` and `PolynomialFeatures`, to identify the most influential factors in predicting high stress levels.

## Prepare Features and Target

### Subtask:
Redefine the `X` (features including 'Sleep_Hours', 'Caffeine_mg', 'Heart_Rate', 'Sleep_Quality', 'Health_Issues') and `y` (target 'Is_High_Stress') DataFrames from the main DataFrame `df`.


```python
features_to_select = ['Sleep_Hours', 'Caffeine_mg', 'Heart_Rate', 'Sleep_Quality', 'Health_Issues']
X = df[features_to_select]
y = df['Is_High_Stress']

print("Features (X) shape:", X.shape)
print("Target (y) shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
print("\nDistribution of target in training set:\n", y_train.value_counts(normalize=True))
print("\nDistribution of target in test set:\n", y_test.value_counts(normalize=True))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline

# Define numerical and categorical features based on the previous analysis and task instructions
numerical_features = ['Sleep_Hours', 'Caffeine_mg', 'Heart_Rate']
categorical_features = ['Sleep_Quality', 'Health_Issues']

# Create a pipeline for numerical features: scaling then polynomial expansion
numerical_transformer = Pipeline(steps=[
    ('scaler', StandardScaler()),
    ('poly', PolynomialFeatures(degree=2, include_bias=False)) # Include bias=False to avoid multicollinearity
])

# Create a transformer for categorical features: one-hot encoding
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

# Combine preprocessing steps using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder='passthrough' # Keep other columns if any, though none are expected here
)

print("Preprocessing steps defined successfully.")
```

```python
from sklearn.linear_model import LogisticRegression

# Create the full pipeline by combining preprocessor and model
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('classifier', LogisticRegression(solver='liblinear', random_state=42))])

print("Machine learning pipeline created successfully, combining preprocessing and Logistic Regression model.")
```

```python
pipeline.fit(X_train, y_train)

print("Pipeline training complete.")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Make predictions on the test set
y_pred = pipeline.predict(X_test)

# Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not High Stress', 'High Stress'], yticklabels=['Not High Stress', 'High Stress'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
classifier = pipeline.named_steps['classifier']
coefficients = classifier.coef_[0]

# Get feature names from the preprocessor
# Numerical feature names after PolynomialFeatures
poly_feature_names = pipeline.named_steps['preprocessor'].named_transformers_['num'].named_steps['poly'].get_feature_names_out(numerical_features)

# Categorical feature names after OneHotEncoder
categ_feature_names = pipeline.named_steps['preprocessor'].named_transformers_['cat'].get_feature_names_out(categorical_features)

# Combine all feature names
all_feature_names = list(poly_feature_names) + list(categ_feature_names)

# Create a DataFrame for feature importance
feature_importance = pd.DataFrame({
    'Feature': all_feature_names,
    'Coefficient': coefficients
})

feature_importance['Absolute_Coefficient'] = abs(feature_importance['Coefficient'])
feature_importance = feature_importance.sort_values(by='Absolute_Coefficient', ascending=False)

print("Feature Importance from Logistic Regression:")
print(feature_importance.head(10))

plt.figure(figsize=(12, 8))
sns.barplot(x='Absolute_Coefficient', y='Feature', data=feature_importance.head(20), palette='viridis')
plt.title('Top 20 Most Influential Factors for High Stress Levels (Absolute Coefficient)')
plt.xlabel('Absolute Coefficient Value')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```
